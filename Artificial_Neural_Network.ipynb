# Artificial Neural Network

This notebook includes: 
* Implementing an Artificial Neural Network to determine customer churn rate at a financial institution
* Encoding categorical data and scaling features
* Inititializing, compiling, and fitting multi-layer Artifical Neural Network to training set
* Creating a Confusion Matrix to visualize results

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix
import sys
import warnings
if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
# Import dataset
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
# Explore dataset
print(dataset.shape)
print(dataset.describe())
dataset.head(10)
```

```python
# Encode categorical data
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Scale features
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Initialize Artificial Neural Network
classifier = Sequential()
```

```python
# Add input layer and first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
```

```python
# Add second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# Add output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compile Artificial Neural Network
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fit Artificial Neural Network to training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Predict test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
```

```python
# Create Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
